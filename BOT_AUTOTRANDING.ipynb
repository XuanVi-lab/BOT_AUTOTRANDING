```python
# Kết nối với Google Drive để lấy dữ liệu
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Import các thư viện
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Đọc dữ liệu từ file CSV trong Google Drive
data = pd.read_csv('/content/drive/My Drive/data.csv')
data.head()
```

```python
# Tính toán trung bình động (moving average)
data['Moving_Average'] = data['Close'].rolling(window=20).mean()
data.head()
```

```python
# Vẽ biểu đồ giá đóng cửa và đường trung bình động
plt.figure(figsize=(10,5))
plt.plot(data['Close'], label='Close Price')
plt.plot(data['Moving_Average'], label='Moving Average')
plt.legend()
plt.show()
```

```python
# Lưu dữ liệu đã xử lý trở lại Google Drive
data.to_csv('/content/drive/My Drive/processed_data.csv', index=False)
```
